```python
import pandas as pd
import numpy as np
import re
import torch

from google.colab import files
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
```

```python
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

df = pd.read_csv(file_name)

# Menampilkan ukuran dataset, contoh data, dan distribusi label
print("Ukuran dataset:", df.shape)
print(df.head())
print("\nDistribusi label awal:")
print(df["label"].value_counts())
```

```python
df["label"] = (
    df["label"]
    .astype(str)
    .str.lower()
    .str.strip()
)

df["label"] = df["label"].replace({
    "hatespeech": "hate speech",
    "hate_speech": "hate speech",
    "hate speech ": "hate speech"
})

valid_labels = ["positive", "neutral", "insult", "hate speech", "threat"]

df = df[df["label"].isin(valid_labels)].copy()

print("Label setelah dirapikan:")
print(df["label"].unique())

print("\nDistribusi label:")
print(df["label"].value_counts())
```

```python
slang_dict = {
    "gk": "tidak",
    "ga": "tidak",
    "gak": "tidak",
    "nggak": "tidak",
    "ngga": "tidak",
    "tdk": "tidak",
    "lu": "kamu",
    "lo": "kamu",
    "gw": "saya",
    "gue": "saya",
    "bgt": "banget",
    "bangett": "banget",
    "yg": "yang",
    "dr": "dari",
    "dri": "dari",
    "sm": "sama",
    "trs": "terus",
    "aja": "saja",
    "udh": "sudah",
    "blm": "belum",
    "org": "orang",
    "anjir": "anjing"
}
```

```python
# PREPROCESSING DATA
def cleaning_text(text):
    text = str(text)
    text = re.sub(r"http\S+|www\S+", " ", text)      # hapus URL
    text = re.sub(r"@\w+", " ", text)                # hapus mention
    text = re.sub(r"#\w+", " ", text)                # hapus hashtag
    text = re.sub(r"[^a-zA-Z0-9\s]", " ", text)      # hapus simbol/emoji
    text = re.sub(r"\s+", " ", text).strip()         # hapus spasi berlebih
    return text

def case_folding(text):
    return text.lower()

def normalize_slang(text):
    words = text.split()
    words = [slang_dict.get(word, word) for word in words]
    return " ".join(words)

def preprocess_text(text):
    text = cleaning_text(text)
    text = case_folding(text)
    text = normalize_slang(text)
    return text

# Terapkan preprocessing ke kolom comment
df["text"] = df["comment"].apply(preprocess_text)

# Cek hasil preprocessing
df[["comment", "text", "label"]].head()
```

```python
def make_stage1_label(label):
    label_str = str(label).strip()
    if label_str in ["insult", "hate speech", "threat"]:
        return "negative"
    return label_str

df["label_tahap1"] = df["label"].apply(make_stage1_label)

df.loc[df["label"].astype(str).str.strip() == "hate speech", "label_tahap1"] = "negative"

# Dataset tahap 2 hanya mengambil data yang termasuk negative
df_negative = df[df["label_tahap1"] == "negative"].copy()
df_negative["label_tahap2"] = df_negative["label"]

print("Distribusi Tahap 1:")
print(df["label_tahap1"].value_counts())

print("\nDistribusi Tahap 2:")
print(df_negative["label_tahap2"].value_counts())
```

```python
# LABEL ENCODING
label_map_tahap1 = {
    "positive": 0,
    "neutral": 1,
    "negative": 2
}

id2label_tahap1 = {
    0: "positive",
    1: "neutral",
    2: "negative"
}

label_map_tahap2 = {
    "insult": 0,
    "hate speech": 1,
    "threat": 2
}

id2label_tahap2 = {
    0: "insult",
    1: "hate speech",
    2: "threat"
}

df["label_tahap1_id"] = df["label_tahap1"].map(label_map_tahap1).astype(pd.Int64Dtype())
df.dropna(subset=["label_tahap1_id"], inplace=True)

df_negative = df[df["label_tahap1"] == "negative"].copy()
df_negative["label_tahap2"] = df_negative["label"]

df_negative["label_tahap2_id"] = df_negative["label_tahap2"].map(label_map_tahap2).astype(pd.Int64Dtype())

df_negative.dropna(subset=["label_tahap2_id"], inplace=True)

print(df[["label", "label_tahap1", "label_tahap1_id"]].head())
print(df_negative[["label", "label_tahap2", "label_tahap2_id"]].head())
```

```python
# TOKENIZER INDOBERT
model_name = "indobenchmark/indobert-base-p1"
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# DATASET CLASS
class TextDataset(torch.utils.data.Dataset):
    def __init__(self, texts, labels):
        self.encodings = tokenizer(
            list(texts),
            truncation=True,
            padding=True,
            max_length=128
        )
        self.labels = list(labels)

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item["labels"] = torch.tensor(self.labels[idx], dtype=torch.long)
        return item

    def __len__(self):
        return len(self.labels)
```

```python
# FUNGSI METRIK EVALUASI
def compute_metrics_manual(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average="macro", zero_division=0)
    recall = recall_score(y_true, y_pred, average="macro", zero_division=0)
    f1 = f1_score(y_true, y_pred, average="macro", zero_division=0)
    return accuracy, precision, recall, f1
```

```python
# FUNGSI TRAINING MODEL
def train_single_model(train_dataset, lr, epoch, num_labels, output_dir):
    model = AutoModelForSequenceClassification.from_pretrained(
        model_name,
        num_labels=num_labels,
        ignore_mismatched_sizes=True
    )

    training_args = TrainingArguments(
        output_dir=output_dir,
        learning_rate=lr,
        per_device_train_batch_size=16,
        per_device_eval_batch_size=16,
        num_train_epochs=epoch,
        save_strategy="epoch", # Changed from "no" to "epoch" to allow saving
        logging_steps=10,
        report_to="none"
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset
    )

    trainer.train()
    trainer.save_model(output_dir) # Explicitly save the model after training
    tokenizer.save_pretrained(output_dir) # Save tokenizer as well
    return trainer
```

```python
# FUNGSI PREDIKSIi
def predict_labels(trainer, dataset):
    outputs = trainer.predict(dataset)
    logits = outputs.predictions
    preds = np.argmax(logits, axis=1)
    return preds
```

```python
# TRAINING 8 PERCOBAAN
# Percobaan dilakukan dengan:
# Split data: 80:20 dan 70:30
# Learning rate: 3e-5 dan 5e-5
# Epoch: 5 dan 8
# Batch size: 16
#
# Total percobaan:
# 2 split x 2 learning rate x 2 epoch = 8 percobaan
#
# Setiap percobaan melatih:
# - model tahap 1
# - model tahap 2
#
# Hasil akhir yang dipakai adalah final_f1_macro.

all_results = []
all_reports = {}

splits = {
    "80:20": 0.2,
    "70:30": 0.3
}

configs = [
    {"lr": 3e-5, "epoch": 5},
    {"lr": 3e-5, "epoch": 8},
    {"lr": 5e-5, "epoch": 5},
    {"lr": 5e-5, "epoch": 8}
]

experiment_id = 1

for split_name, test_size in splits.items():

    # Split data untuk tahap 1
    train_df, test_df = train_test_split(
        df,
        test_size=test_size,
        random_state=42,
        stratify=df["label_tahap1_id"]
    )

    # Split data untuk tahap 2
    train_neg_df, test_neg_df = train_test_split(
        df_negative,
        test_size=test_size,
        random_state=42,
        stratify=df_negative["label_tahap2_id"]
    )

    train_stage1_dataset = TextDataset(train_df["text"], train_df["label_tahap1_id"])
    test_stage1_dataset = TextDataset(test_df["text"], test_df["label_tahap1_id"])

    train_stage2_dataset = TextDataset(train_neg_df["text"], train_neg_df["label_tahap2_id"])
    test_stage2_dataset = TextDataset(test_neg_df["text"], test_neg_df["label_tahap2_id"])

    for config in configs:
        lr = config["lr"]
        epoch = config["epoch"]

        print("="*80)
        print(f"EKSPERIMEN {experiment_id}")
        print(f"Split         : {split_name}")
        print(f"Learning Rate : {lr}")
        print(f"Epoch         : {epoch}")
        print(f"Batch Size    : 16")
        print("="*80)

        # ===============================
        # Training Tahap 1
        # ===============================
        print("Training Tahap 1: positive / neutral / negative")

        trainer_stage1 = train_single_model(
            train_stage1_dataset,
            lr,
            epoch,
            num_labels=3,
            output_dir=f"./exp_{experiment_id}_stage1"
        )

        pred_stage1 = predict_labels(trainer_stage1, test_stage1_dataset)
        true_stage1 = test_df["label_tahap1_id"].values

        acc1, prec1, rec1, f1_1 = compute_metrics_manual(true_stage1, pred_stage1)

        report_stage1 = classification_report(
            true_stage1,
            pred_stage1,
            target_names=["positive", "neutral", "negative"],
            zero_division=0
        )

        print("\nREPORT TAHAP 1")
        print(report_stage1)

        # ===============================
        # Training Tahap 2
        # ===============================
        print("Training Tahap 2: insult / hate speech / threat")

        trainer_stage2 = train_single_model(
            train_stage2_dataset,
            lr,
            epoch,
            num_labels=3,
            output_dir=f"./exp_{experiment_id}_stage2"
        )

        pred_stage2 = predict_labels(trainer_stage2, test_stage2_dataset)
        true_stage2 = test_neg_df["label_tahap2_id"].values

        acc2, prec2, rec2, f1_2 = compute_metrics_manual(true_stage2, pred_stage2)

        report_stage2 = classification_report(
            true_stage2,
            pred_stage2,
            target_names=["insult", "hate speech", "threat"],
            zero_division=0
        )

        print("\nREPORT TAHAP 2")
        print(report_stage2)

        # ===============================
        # Evaluasi End-to-End
        # ===============================
        # Jika tahap 1 memprediksi positive/neutral, hasil langsung dipakai.
        # Jika tahap 1 memprediksi negative, maka teks diproses lagi oleh model tahap 2.

        final_true = list(test_df["label"])
        final_pred = []

        for text, stage1_pred in zip(test_df["text"], pred_stage1):
            stage1_label = id2label_tahap1[stage1_pred]

            if stage1_label in ["positive", "neutral"]:
                final_pred.append(stage1_label)
            else:
                temp_dataset = TextDataset([text], [0])
                stage2_pred = predict_labels(trainer_stage2, temp_dataset)[0]
                final_pred.append(id2label_tahap2[stage2_pred])

        final_labels = ["positive", "neutral", "insult", "hate speech", "threat"]

        acc_final = accuracy_score(final_true, final_pred)
        prec_final = precision_score(final_true, final_pred, labels=final_labels, average="macro", zero_division=0)
        rec_final = recall_score(final_true, final_pred, labels=final_labels, average="macro", zero_division=0)
        f1_final = f1_score(final_true, final_pred, labels=final_labels, average="macro", zero_division=0)

        report_final = classification_report(
            final_true,
            final_pred,
            labels=final_labels,
            zero_division=0
        )

        print("\nREPORT END-TO-END")
        print(report_final)

        all_results.append({
            "experiment": experiment_id,
            "split": split_name,
            "learning_rate": lr,
            "epoch": epoch,
            "batch_size": 16,

            "stage1_accuracy": acc1,
            "stage1_precision_macro": prec1,
            "stage1_recall_macro": rec1,
            "stage1_f1_macro": f1_1,

            "stage2_accuracy": acc2,
            "stage2_precision_macro": prec2,
            "stage2_recall_macro": rec2,
            "stage2_f1_macro": f1_2,

            "final_accuracy": acc_final,
            "final_precision_macro": prec_final,
            "final_recall_macro": rec_final,
            "final_f1_macro": f1_final
        })

        all_reports[f"experiment_{experiment_id}_stage1"] = report_stage1
        all_reports[f"experiment_{experiment_id}_stage2"] = report_stage2
        all_reports[f"experiment_{experiment_id}_final"] = report_final

        experiment_id += 1
```

```python
df_results = pd.DataFrame(all_results)
df_results
```

```python
# SIMPAN HASIL EKSPERIMEN
df_results.to_csv("hasil_pengujian_semua_model.csv", index=False)

with open("classification_reports_semua_model.txt", "w") as f:
    for key, value in all_reports.items():
        f.write("="*80 + "\n")
        f.write(key + "\n")
        f.write("="*80 + "\n")
        f.write(value + "\n\n")

files.download("hasil_pengujian_semua_model.csv")
files.download("classification_reports_semua_model.txt")
```
